DATA PANEN

```python
data_panen = {
    'lokasi1': {
        'nama_lokasi': 'Kebun A',
        'hasil_panen': {
            'padi': 1200,
            'jagung': 800,
            'kedelai': 500
        }
    },
    'lokasi2': {
        'nama_lokasi': 'Kebun B',
        'hasil_panen': {
            'padi': 1500,
            'jagung': 900,
            'kedelai': 450
        }
    },
    'lokasi3': {
        'nama_lokasi': 'Kebun C',
        'hasil_panen': {
            'padi': 1100,
            'jagung': 750,
            'kedelai': 600
        }
    },
    'lokasi4': {
        'nama_lokasi': 'Kebun D',
        'hasil_panen': {
            'padi': 1300,
            'jagung': 850,
            'kedelai': 550
        }
    },
    'lokasi5': {
        'nama_lokasi': 'Kebun E',
        'hasil_panen': {
            'padi': 1400,
            'jagung': 950,
            'kedelai': 480
        }
    }
}
```

Tampilkan seluruh data dari data_panen termasuk nama lokasi dan hasil panen

```python
for lokasi, data in data_panen.items():
    print(f"\n{lokasi.upper()}:")
    print(f"Nama Lokasi: {data['nama_lokasi']}")
    print(f"Hasil Panen:")
    for tanaman, jumlah in data['hasil_panen'].items():
        print(f"{tanaman.capitalize()}: {jumlah}")
```

Tampilkan jumlah hasil panen jagung dari lokasi2

```python
jagung_lokasi2 = data_panen['lokasi2']['hasil_panen']['jagung']
print(f"Jumlah hasil panen jagung dari lokasi2: {jagung_lokasi2}")
```

Tampilkan nama lokasi dari lokasi3

```python
nama_lokasi3 = data_panen['lokasi3']['nama_lokasi']
print(f"Nama lokasi3: {nama_lokasi3}")
```

Masukkan jumlah hasil panen padi dan kedelai setiap lokasi ke variabel terpisah

```python
padi_lokasi1 = data_panen['lokasi1']['hasil_panen']['padi']
kedelai_lokasi1 = data_panen['lokasi1']['hasil_panen']['kedelai']

padi_lokasi2 = data_panen['lokasi2']['hasil_panen']['padi']
kedelai_lokasi2 = data_panen['lokasi2']['hasil_panen']['kedelai']

padi_lokasi3 = data_panen['lokasi3']['hasil_panen']['padi']
kedelai_lokasi3 = data_panen['lokasi3']['hasil_panen']['kedelai']

padi_lokasi4 = data_panen['lokasi4']['hasil_panen']['padi']
kedelai_lokasi4 = data_panen['lokasi4']['hasil_panen']['kedelai']

padi_lokasi5 = data_panen['lokasi5']['hasil_panen']['padi']
kedelai_lokasi5 = data_panen['lokasi5']['hasil_panen']['kedelai']

print(f"Lokasi1 - Padi: {padi_lokasi1}, Kedelai: {kedelai_lokasi1}")
print(f"Lokasi2 - Padi: {padi_lokasi2}, Kedelai: {kedelai_lokasi2}")
print(f"Lokasi3 - Padi: {padi_lokasi3}, Kedelai: {kedelai_lokasi3}")
print(f"Lokasi4 - Padi: {padi_lokasi4}, Kedelai: {kedelai_lokasi4}")
print(f"Lokasi5 - Padi: {padi_lokasi5}, Kedelai: {kedelai_lokasi5}")
```

Cek kondisi setiap lokasi

```python
for lokasi, data in data_panen.items():
    nama = data['nama_lokasi']
    padi = data['hasil_panen']['padi']
    jagung = data['hasil_panen']['jagung']
    
    print(f"\n{lokasi.upper()} ({nama}):")
    print(f"Padi: {padi},Jagung: {jagung}")
    
    if padi > 1300 or jagung > 800:
        print(f"MEMERLUKAN PERHATIAN KHUSUS")
        if padi > 1300:
            print(f"Hasil panen padi melebihi 1300")
        if jagung > 800:
            print(f"Hasil panen jagung melebihi 800")
    else:
        print(f"KONDISI BAIK")

print(f"selesai")
```
